3. Generate random samples from two specified distributions and plot their histograms, side by side in a single figure. Each sub-figure will have its own labeled axes and title.

```python
#np.random.normal(loc=μ, scale=σ, size=...) 但如果是默认的mean 0，sd 1的话就不用，直接size就行。
import numpy as np
import matplotlib.pyplot as plt

n=10000

# Normal(μ = -2, σ = 6)
normal_samples = np.random.normal(loc=-2, scale=6, size=n)

# Chi-Square(k = 4)
chisqr_samples = np.random.chisquare(df=4, size=n)

# Uniform(a = -3, b = 5)
unif_samples = np.random.uniform(low=-3, high=5, size=n)

# nrows=1：一行,ncols=3：三列, 整张图宽9英寸，高3英寸，如果正方形排列看lecture8with solutions 10.2✖️2的话00是左上。
fig, subfig_array = plt.subplots(nrows=1, ncols=3, figsize=(9,3))
#图1
subfig_array[0].hist(x=normal_samples, bins=30, color='C0')
subfig_array[0].set_title("Normal Distribution\n(mean = -2, s.d. = 6)")
subfig_array[0].set_xlabel("Sample Value")
subfig_array[0].set_ylabel("Frequency")
#图2
subfig_array[1].hist(x=unif_samples, bins=30, color='C1')
subfig_array[1].set_title("Uniform Distribution\nrange [-3, 5]")
subfig_array[1].set_xlabel("Sample Value")
subfig_array[1].set_ylabel("Frequency")
#图4
subfig_array[2].hist(x=chisqr_samples, bins=30, color='C2')
subfig_array[2].set_title("Chi-Square Distribution\n(df = 4)")
subfig_array[2].set_xlabel("Sample Value")
subfig_array[2].set_ylabel("Frequency")
#最后，如果前面要加x，y的区间 统一还有不同sample size复杂一点的simulation 看lecture8solution的15。简单的simulation看13。
plt.tight_layout()
plt.show()
```

5. Read in a .csv file to create a Pandas DataFrame. Loop over a list of variable/column names, printing the mean of each variable.



```python
import pandas as pd                      # 导入 pandas 库（数据分析常用工具）
import matplotlib.pyplot as plt          # 导入 matplotlib 库用于画图

# 1️⃣ 从文件夹 practice_quiz_data 读取 CO₂ 排放数据
# 注意：路径要与你的文件位置一致 /就是一层一层
us_emissions = pd.read_csv("practice_quiz_data/united_states_co2_emissions.csv")

# 2️⃣ 检查前几行，确认数据是否正确读入，us_emissions.describe()也可以看到
print(us_emissions.head())

columns = ["year", "total_emissions"] #create a list of variable names，为后面的 for 循环（计算平均值或画图）做准备。方便访问 其实就是data里面的variable
#如果要看crosstab分类就是lecture1的supplement
#print() # prints a blank line
#len看长度或者多少个

# 用 for 循环依次计算每个变量的平均值
for col in columns:
    # 先确认该列是否存在于 DataFrame 中，防止拼写错误
    if col in us_emissions.columns:
        mean_val = us_emissions[col].mean()   # pandas 的 mean() 函数计算平均值
        print("Mean of", col, "=", mean_val)
    else:
        print("Column", col, "not found in DataFrame")
```

6. Create a list of .csv file names and an additional empty list. Then in a for loop, read in each file to make a DataFrame, compute the mean of one of the variables/features/columns, and store it as an element of the originally empty list. Then print the list. 



```python
import pandas as pd   # 用 pandas 读取 csv 文件，并创建 DataFrame

# Create a list of .csv file names ，手动建立一个列表，里面是每个国家的 CO2 数据文件路径
file_list = [
    "practice_quiz_data/china_co2_emissions.csv",
    "practice_quiz_data/germany_co2_emissions.csv",
    "practice_quiz_data/india_co2_emissions.csv",
    "practice_quiz_data/indonesia_co2_emissions.csv",
    "practice_quiz_data/iran_islamic_rep_co2_emissions.csv",
    "practice_quiz_data/japan_co2_emissions.csv",
    "practice_quiz_data/korea_rep_co2_emissions.csv",
    "practice_quiz_data/russian_federation_co2_emissions.csv",
    "practice_quiz_data/saudi_arabia_co2_emissions.csv",
    "practice_quiz_data/united_states_co2_emissions.csv"
]

# Create an empty list 用来存储每个文件的 total_emissions 平均值
mean_list = []

# For-loop: read in each file and compute mean 
for filename in file_list:                           # 依次处理每一个国家的 CSV 文件
    df = pd.read_csv(filename)                       # 读取 CSV 文件为 DataFrame，pd.read_csv() 。
    mean_value = df["total_emissions"].mean()        # 计算 total_emissions 列的平均值
    #mean_value = round(mean_value, 2) - 保留两位小数，结果更清晰，可以不管
    mean_list.append(mean_value)  # 把结果添加进空列表 mean_list，这一行完成了“store it as an element of the list”
    # 打印出来np.float什么的不好看的话可以 mean_list.append(float(mean_value))

    # 如果要检查的话就每次循环都打印结果，这是循环内的
    # print("File processed:", filename)
    # print("Mean of total_emissions:", mean_value)
    # print("-" * 50) 这是分割线可要可不要

# Print the final list， 把所有国家的平均值都打印出来，这是循环外的
print("Final list of average total_emissions:")
print(mean_list)
```
